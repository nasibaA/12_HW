# Standardized data and modeling

K -nearest neighbors is a model that classifies data based on its distance to training set data.

Before training your model it is very important to split up your data into traingn and test sets, to avoid overfitting.

- most models have a score function which can be used to evaluate your model's performance

K-nearest neighbors (KNN) is a type of supervised learning algorithm used for both regression and classification. KNN tries to predict the correct class for the test data by calculating the distance between the test data and all the training points.

```python
def fill_missing_values(df):
     for col in df.select_dtypes(include= [int,float]).columns:
             val = df[col].mean()
                 df[col].fillna(val, inplace=True) ## replace the missing values in the numerical columns with the mean value of the column
                     return df
```

```python
def drop_duplicates(df, column_name):
     df = df.drop_duplicates(subset=column_name)
         return df # removing the duplicate values
```

```python
def remove_outliers(df, column_list):
     for col in column_list:
             avg = df[col].mean()
                 std = df[col].std() ## eliminating the outliers
                     low = avg  2 * std
                         high = avg + 2 * std
                             df = df[df[col].between(low, high, inclusive=True)]
                                    return df
```
